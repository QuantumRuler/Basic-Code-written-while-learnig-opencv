```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(img):
    rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(rgb)
    plt.show()
```

```python
img = cv.imread('./Photos/cat.jpg')
```

```python
#BGR to gray
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
#BGR to HSV
hsv = cv.cvtColor(img,cv.COLOR_BGR2HSV)
```

```python
#BGR to L+a+b
lab = cv.cvtColor(img,cv.COLOR_BGR2LAB)
```

```python
plt.imshow(img)
plt.show()
```

```python
rgb_img = cv.cvtColor(img , cv.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
```

```python
## Blur
```

```python
#Average
average = cv.blur(img,(5,5))
plot(average)
```

```python
#Gaussian blur
#it you can add std dev in or y direction
gblur = cv.GaussianBlur(img, (7,7),0)
plot(gblur)
```

```python
#median
median = cv.medianBlur(img,3)
plot(median)
```

```python
#manually adding avg
kenrnal = np.ones((5,5),np.float32)/25
m_average = cv.filter2D(img,-1,kenrnal)
plot(average)
plot(m_average)
```

```python
#bilatral
# usally used in adv cv project to redeuce noise
#you blur the o=img nbut you retain the edges
bilateral = cv.bilateralFilter(img,7,15,35)
plot(bilateral)
```
